```python
# you should not import any other packages
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
import numpy as np
from sklearn.linear_model import SGDRegressor
```

```python
import numpy as np
import scipy as sp
import scipy.optimize

def angles_in_ellipse(num,a,b):
    assert(num > 0)
    assert(a < b)
    angles = 2 * np.pi * np.arange(num) / num
    if a != b:
        e = (1.0 - a ** 2.0 / b ** 2.0) ** 0.5
        tot_size = sp.special.ellipeinc(2.0 * np.pi, e)
        arc_size = tot_size / num
        arcs = np.arange(num) * arc_size
        res = sp.optimize.root(
            lambda x: (sp.special.ellipeinc(x, e) - arcs), angles)
        angles = res.x 
    return angles
```

```python
a = 2
b = 9
n = 50

phi = angles_in_ellipse(n, a, b)
e = (1.0 - a ** 2.0 / b ** 2.0) ** 0.5
arcs = sp.special.ellipeinc(phi, e)

fig = plt.figure()
ax = fig.gca()
ax.axes.set_aspect('equal')
ax.scatter(b * np.sin(phi), a * np.cos(phi))
plt.show()
```

```python
## referred https://stackoverflow.com/questions/59367939/how-to-correct-the-position-of-hyper-plane-in-python for optimization

c = 1
alpha = [0.0001, 1, 100]
outliers = [(0,2),(21, 13), (-23, -15), (22,14), (23, 14)]
plt.figure(figsize=(20,15))
for i,a1 in enumerate(alpha):
    X1= b * np.sin(phi)
    Y= a * np.cos(phi)
    for j,o in enumerate(outliers):
        plt.subplot(3,5,c)
        c += 1
        X1 = np.append(X1, o[0]).reshape(-1,1)     ## adding the outliers
        Y = np.append(Y, o[1]).reshape(-1,1)
        sgdr = SGDRegressor(alpha = a1, eta0=0.0001, max_iter=1000, learning_rate='invscaling',random_state=15, tol=1e-3)
        sgdr.fit(X1,Y)
        Y_pred = sgdr.predict(X1)
        plt.scatter(X1,Y)
        plt.plot(X1, Y_pred, color='red')  ## plotting the hyperplane
        plt.title('alpha = '+ str(a1))
        
print("sgd regressor")
plt.show()
```

### Observation:

* When there is only one outlier, the datapoints are fairly spread and the hyperplane passes from between the points horizontally.
* As number of outliers increase for low alpha value, the hyperplane changes its inclination to cover all the points.
* As number of outliers increase for alpha value of 1, the hyperplane position is similar to one when alpha = 0.0001
* And as outliers increase, the hyperplane adjusts itself slightly to include all the points.
* The position of hyperplane in alpha 100 is better than the previous ones.
* It can be concluded that when alpha is too low, hyperplane is much inclined towards the outliers. As alpha and outliers increase, there is a slight change in hyperplane position but it fits better than the previous ones.
